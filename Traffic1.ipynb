```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df=pd.read_csv("accident data.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df2=df.fillna("unknown")
```

```python
df2
```

```python
df2.isnull().sum()
```

```python
df2.duplicated().sum()
```

```python
df3=df2.drop_duplicates()
```

```python
df3.duplicated().sum()
```

```python
df.columns
```

```python
acc_severity=df['Accident_Severity'].value_counts()
```

```python
plt.figure(figsize=(8,8))
plt.pie(acc_severity ,labels=acc_severity.index,startangle=140)
plt.title('Accident Severity Distribution')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Road_Surface_Conditions' , data=df3 , palette='coolwarm')
plt.title("Accidents by Road Surface Condition")
plt.xlabel('Road Surface Condition')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
df['Accident Date']=pd.to_datetime(df['Accident Date'],dayfirst=True)
```

```python
df['Month']=df['Accident Date'].dt.month_name()
df['Day']=df['Accident Date'].dt.day
```

```python
df3.head()
```

```python
data=df3.pivot_table(index='Day',columns='Month',aggfunc='size',fill_value=0)
plt.figure(figsize=(15,6))
sns.heatmap(data,annot=True,fmt='d',cmap='coolwarm')
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df3,y='Weather_Conditions',palette='coolwarm')
plt.title('Accidents by Weather Conditions')
plt.show()
```
